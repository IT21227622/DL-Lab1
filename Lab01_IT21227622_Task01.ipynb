```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title('3D Plot of Z = X^2 + Y^2')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```

```python
import pandas as pd

# Load the dataset
data = pd.read_csv('Pokemon.csv', encoding='ISO-8859-1')
features = data[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]

# Calculate Pearson and Spearman correlations
pearson_corr = features.corr(method='pearson')
spearman_corr = features.corr(method='spearman')

# Visualize the correlations using heatmaps
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize=(14, 6))
sns.heatmap(pearson_corr, annot=True, ax=ax[0], cmap='coolwarm')
ax[0].set_title('Pearson Correlation')
sns.heatmap(spearman_corr, annot=True, ax=ax[1], cmap='coolwarm')
ax[1].set_title('Spearman Correlation')
plt.show()
```
