```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Au_nanoparticle_dataset.csv')
```

```python
new_df = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]
```

```python
print(new_df.head(20))
```

```python
mean_values = new_df.mean()
print(mean_values)
```

```python
std_values = new_df.std()
print(std_values)
```

```python
quartiles = new_df.quantile([0.25, 0.5, 0.75])
print(quartiles)
```

```python
import matplotlib
```

```python
plt.figure(figsize=(20, 5))

for i, column in enumerate(new_df.columns):
    plt.subplot(1, 4, i + 1)
    plt.hist(new_df[column], bins=30, alpha=0.7, color='purple')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

sns.pairplot(new_df)
plt.show()
```

```python
g = sns.PairGrid(new_df)
g.map_upper(sns.histplot, bins=30)
g.map_diag(sns.histplot, kde=True)
g.map_lower(sns.kdeplot)

plt.show()
```
